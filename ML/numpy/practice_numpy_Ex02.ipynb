```python
num = int(input("Please Enter Your Number: "))
df = np.array([np.random.randint(100) for i in range(num)])
print(df)

max_number = np.max(df)
print(max_number)
if max_number > 70:
    print(f"{max_number} is greater than 70 --> win")
else:
    print(f"{max_number} is smaller than 70 --> Fail")
```

```python
import numpy as np
from numpy import random


number = int(input ('Please enter your number: '))
x = random.choice(100, size=number)
print(x)

max_number = np.max(x)
print(max_number)
if max_number >70:
    print('WIN')
else:
    print('FAIL')
```
