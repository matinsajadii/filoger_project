```python
text = input("Enter a text: ").replace(" ", "")
freq = {}
for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

      
print(f"{'+-------+------------+'}\n| NAME\t| FREQUENCY  |\n{'+=======+============+'}")
for key, val in freq.items():
  print(f"|{' '*3}{key}\t|{' '*5}{val}\t{' '*5}|")
  print("+" + "-"*7 + "+" + "-"*12 + "+")

# print(f'+-------+------------+')
```

```python
# make it suitable for caseless comparisions
text = input("Enter Your Text: ").casefold()

# convert to dict
char_dict = dict.fromkeys(text,0)

#count the input text
for char in text:
  if char in char_dict:
    char_dict[char] += 1

print(f"{'+-------+------------+'}\n| NAME\t| FREQUENCY  |\n{'+-------+------------+'}")
for key, val in char_dict.items():
  print(f"|{' '*3}{key}\t|{' '*5}{val}\t{' '*5}|")

print(f'+-------+------------+')
```

```python
from prettytable import PrettyTable
text = input("Enter Your Text: ").casefold()
table = PrettyTable(["Name", "Frequency"])
for key, val in dict.fromkeys(text,0).items():
  table.add_row([key, val])

print(table)
```

```python
from prettytable import PrettyTable
from collections import Counter
text = Counter(input("Enter Your Text: ").casefold())
t = PrettyTable(["Name", "Frequency"])
t = t.add_row({key:val for key, val in text.items()})

print(t)
```

```python
from prettytable import PrettyTable
from collections import Counter
text = Counter(input("Enter Your Text: ").lower().replace(" ", ""))
table = PrettyTable(['Name', 'Frequency'])
for list_ in [[key,val] for key, val in text.items()]:
  table.add_row(list_)
print(table)
```
